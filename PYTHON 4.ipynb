## **Functions and Methods**

- **FUNCTION:** indipendent block of code.

```python
# example:
print("great")
```

- **METHOD:** this is a function but belongs to object or a class.

```python
# example:
"great".upper()
```

`Function`

- Exists independently
- Called directly by name
- Doesn't automatically receive object information
- Example: `add(2, 3)`

`Method`

- Belongs to a class/object
-  Called through an object or class
-  Usually receives the instance (self) or class (cls) automatically
-  Example: calc.`add(2, 3)`

**Short definition:**

* **Function:** Called directly by its name.
* **Method:** Called using **dot (`.`) notation** on an object or class.

#### `1. capitalize()`

- capitalize() returns a new string with the first character in uppercase and all remaining characters in lowercase.

```python
text = "python"
```

```python
text.capitalize()
```

#### `2. casefold()`

- casefold() returns a new string with all alphabetic characters converted to lowercase. It is more aggressive than lower() and is mainly used for case-insensitive comparisons.

```python
animal = "ZebRA"
```

```python
animal.casefold() #lower also do same but casefold do this work for all language possibly strictly.
```

- returns a new string centered within a specified width. If needed, it adds a fill character (default is a space) on both sides.

```python
text = "Python"
print(text.center(100))
```

```python
text = "Python"
print(text.center(100,"-"))
```

#### `3. count()`

- count() returns the number of times a substring appears in a string.

string.count(substring, start, end)

```python
text = "banana"
```

```python
text.count("a")
```

#### `4.encode()`

- encode() converts a string into a bytes object using a specified encoding (default is UTF-8).

```python
text = "Python"
b = text.encode()
print(b) # here b represents that bytes
```

#### 5. `endswith()`

- endswith() returns True if a string ends with the specified suffix; otherwise, it returns False

```python
text = "Hello World"
```

```python
print(text.endswith("World"))
```

```python
print(text.endswith("World", 6)) # first it checks (6:) which is world then check endwith.
```

```python
print(text.endswith("world", 0, 5)) # here hello is suffix 0 is start 5 is end
# first its check (0:5) which is hello then check endswith world so obviously its no match.
```

#### `6. expandtabs()`

- expandtabs() replaces every tab character (\t) in a string with spaces.

```python
text = "Hello\tWorld"
text.expandtabs()
```

```python
print(repr(text)) # repr() displays an object the way Python internally represents it.
print(repr(text.expandtabs()))
```

#### `7. find()`

- find() returns the index of the first occurrence of a substring. If the substring is not found, it returns -1.

```python
text = "Python Programming"  

text.find("Program") # "Program" starts at index 7.
```

```python
text = "banana"

text.find("a") # note they tells the position exactly first time matching.
```

```python
text = "banana"

print(text.find("x")) # not found
```

```python
text = "banana"

print(text.find("a", 2, 5))
```

```python
# important point
# find() → returns -1 if not found.
# index() → raises ValueError if not found.
```

#### `8. format()`

- format() is used to insert values into a string.
- format() inserts values into placeholders {} and returns a formatted string.

```python
text = "My name is {}."

print(text.format("siddiqui"))
```

```python
text = "My name is {} and I am {} years old."

print(text.format("siddiqui", 20))
```

```python
text = "{0} is older than {1}"

print(text.format("siddiqui", "Ahmed"))
```

```python
text = "Name: {name}, Age: {age}"

print(text.format(name="siddiqui", age=20))
```

#### `9. format_map()`

- format_map() formats a string using the values from a dictionary.

```python
data = {"name": "Ali", "age": 20}

text = "Name: {name}, Age: {age}"

print(text.format_map(data))
```

```python
student = {
    "name": "Abdullah",
    "course": "Python"
}

text = "{name} is learning {course}."

print(text.format_map(student))
```

- format_map() fills placeholders using a dictionary.

#### `10. index()`

- index() returns the index of the first occurrence of a substring. If the substring is not found, it raises a ValueError.

```python
text = "banana"

print(text.index("a"))
```

```python
text = "banana"

print(text.index("a", 2)) # Python starts searching from index 2 but rule doesnot break.
```

```python
text = "banana"

print(text.index("x"))
```

#### `12.isalnum()`

- isalnum() returns True if all characters in the string are alphanumeric (letters or digits) and the string is not empty. Otherwise, it returns False

```python
text = "Python"

print(text.isalnum())
```

```python
text = "Python 123"

print(text.isalnum()) # A space is not an alphanumeric character
```

```python
text = "Python@123"

print(text.isalnum()) #@ is a special character
```

`What is "Alphanumeric"?`

`Alphanumeric = Alphabet + Numeric`

`Allowed:`

- A-Z
- a-z
- 0-9

`Not allowed:`
- any special character

#### `14. isascii()`

- isascii() returns True if all characters in the string are ASCII characters (0–127), or if the string is empty. Otherwise, it returns False.

```python
text = "Python"

print(text.isascii())
```

```python
text = "Python123"

print(text.isascii())
```

```python
text = "Hello@123"

print(text.isascii())
```

```python
text = "20₹cafe"

print(text.isascii())
```

#### `15. isdecimal()`

- isdecimal() returns True if all characters in the string are decimal digits (0–9) and the string is not empty. Otherwise, it returns False.

```python
text = "123abc"

print(text.isdecimal())
```

```python
text = "123"

print(text.isdecimal())
```

```python
text = "12.3"

print(text.isdecimal()) # float not allowed
```

#### `15. isdigit()`

- isdigit() returns True if all characters in the string are digits and the string is not empty. Otherwise, it returns False

```python
text = "12345"

print(text.isdigit())
```

```python
text = "12 34"

print(text.isdigit()) # space not allowed
```

```python
text = "12.34"

print(text.isdigit()) # float not allowed
```

The difference is that isdigit() accepts some extra Unicode digit characters, while isdecimal() is stricter

```python
text = "²"

print(text.isdecimal())
print(text.isdigit())
```

```python
# Here, ² (superscript two) is considered a digit, but not a decimal digit.
```
