#### `16. isidentifier()`

- isidentifier() returns True if the string is a valid Python identifier; otherwise, it returns False.

- An identifier can start with a letter or _  and after that it can contain letters, digits, and _ only otherwise its gets false.

```python
print("name".isidentifier())      # True
print("_name".isidentifier())     # True
print("name123".isidentifier())   # True
print("123name".isidentifier())   # False its starts with numbers.
print("my name".isidentifier())   # False its having dash inside it.
print("myname@".isidentifier())   # it contains special character.
```

#### `19.isprintable()`

- isprintable() returns True if all characters in the string are printable or if the string is empty. Otherwise, it returns False.

**What are printable characters?**

**Printable characters are characters that can be displayed on the screen, such as:**

-  Letters (A-Z, a-z)
-  Digits (0-9)
-  Spaces
-  Symbols (@, #, !, $, etc.)

**Non-printable characters include:**

-  \n (newline)
-  \t (tab)
-  \r (carriage return)

- A non-printable character is a character that cannot be displayed as a visible symbol on the screen. Instead, it performs a special action.

```python
text = "Hello\tWorld"

print(text.isprintable())
```

```python
text = "helloworld"
print((text).isprintable())
```

```python
text = "helloworld"
print((text).isprintable())
```

```python
text = "Python123"

print(text.isprintable())
```

- \r moves the cursor to the beginning of the current line without moving to a new line.

```python
print("Hello\rHi") # Moves to the beginning of the same line.
```

```python
print("12345\rAB")
```

#### `23.join()`

- join() joins the elements of an iterable into a single string using the specified separator.

- separator → The string placed between the elements.
- iterable → A list, tuple, set, etc., containing strings.

```python
words = ["Python", "is", "awesome"] # A space (" ") is inserted between each word.

print(" ".join(words))
```

```python
fruits = ["Apple", "Banana", "Mango"] # comma is added between every word.

print(", ".join(fruits))
```

```python
letters = ["A", "B", "C"] # Hyphen is added between them.

print("-".join(letters))
```

```python
letters = ["P", "Y", "T", "H", "O", "N"] # No separator is added.

print("".join(letters))
```

```python
colour = [ "orange" , "indigo" , "grey"]
print("<".join(colour))
```

```python
data = ["Ali", 20] # this gets error because  20 is int not str.

print(", ".join(data))
```
