## **NATURAL LANGUAGE**

**Natural Language:**
A natural language is a human language used for everyday communication, such as English, Hindi, Arabic. It is spoken and written naturally by people.


## **HIGH LEVEL LANGUAGE**

**High-Level Language:**
A high-level language is a programming language that is easy for humans to read, write, and understand, such as Python, Java, and C++. It is closer to human language than machine language.


## **LOW LEVEL LANGUAGE**

**Low-Level Language:**
A low-level language is a programming language that is closer to machine language and hardware. It is difficult for humans to read and write but provides greater control over the computer. Examples include Assembly Language and Machine Language.


## **ASSEMBLY LANGUAGE**

**Assembly Language:**
Assembly language is a low-level programming language that uses mnemonic codes and symbols instead of binary numbers, making it easier to understand than machine language.

**Examples:** `MOV A, B`, `ADD A, 10`, `SUB B, A`.


## **MACHINE LANGUAGE**

**Machine Language:**
Machine language is the lowest-level programming language made up of binary digits (0s and 1s) that a computer can understand and execute directly.

**Example:** `10110000 01100001` (binary code).

## **PYTHON DEFINITION**

**Python is both a compiled and interpreted language. First, the Python compiler translates the source code into bytecode, which is an intermediate form of code. Then, the Python Virtual Machine (PVM) executes the bytecode line by line, converting it into machine-level instructions (binary form), allowing the computer to run the program and produce the desired output.**

HERE **Bitecode** means **Bytecode is an intermediate set of instructions generated by the Python compiler, which is executed by the Python Virtual Machine (PVM). For example, the Python statement `print("Hello")` may be converted into bytecode instructions such as `LOAD_CONST "Hello"` and `PRINT_EXPR` before execution.**


**Python was created by Guido van Rossum in the late 1980s, and its first public release was in 1991.**
