## **ESCAPE SEQUENCES**

- `\" = BACKSLASH WITH DOUBLE QUOTE # allows you to include a " without ending it.`
- `\' = BACKSLASH WITH SINGLE QUOTE # allows you to include a ' without ending it.`
- `\\ = DOUBLE BACKSLASH # allows one \ as a output result without actually its works.`
- `\t = BACKSLASH WITH #it creates a tab space` 
- `\n = bachslash with n # it creates new line and moves the text that comes after it to the next - line.`
- `\b = BACKSLASH WITH b # this is a backspace escape sequence that moves the cursor one character backward and allows the next character to replace the previous one.`

```python
print("universal truth\"yes\"ok")
```

```python
print('universal \'morality\'ok')
```

```python
print("hello\\noble\\x")
```

```python
print ("life is \t awesome")
```

```python
print("This is \npython language")
```

```python
print("123\b45")
```
