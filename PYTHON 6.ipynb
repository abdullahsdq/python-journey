## **f-string**

- modern and easy way to format and build strings "f" stands for "formatted".

- f string makes concatenate with a help of variables.

```python
name = "abdullah"
age = 20
student = True

print("My name is " + name + " my age is " + str(age) + " and I'm a student: " + str(student))
# this makes works slow and typical by using str ,int.
```

```python
name = "abdullah"
age = 20
student = True

print(f"My name is + {name} my age is {age} and I'm a student: {student}" )
# f string makes concatenate easy with a help of curly brakets {}.
```

- f string makes expression easy in calculation .

```python
print(f"20+80 = {20+80}") # this is expression  working without variable.
```

```python
print(f"{this is me}") 
# this is not working because here inside{} this is not variable nor expression.
```

```python
print(f"{{this is me siddiqui 20y old }}") 
# if you write forcefully you will add one more curly brackets {{}}.
```
