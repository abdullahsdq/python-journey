#### for short remembering datatypes concepts

- **string**: is the datatype using double quotes and single quotes from both side.
- **int**: is the numeric datatype using numbers.
- **float**: is the decimal datatype using numbers and point.
- **list**: is have ability to containing all the datatype in one variable using square brackets[]
- **tuple**: also have same ability but using using parenthesis()
- **set**: also have same ability but using curly brackets{}
- **dictionary**: is contain keys and valus and makes by using curly brackets and colon { : }
- **None**: means no value ,nothing,uses to show absence of any data.
- **bool** : is True ,False datatype.
